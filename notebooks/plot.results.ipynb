## Get coordinates of local maxima

From the text file that is part of the nipype clustering routine output.

```python
# get coordinates of local maxima from file
def get_localmaxima(textfile):
    """
    Reads csv-file with local maxima (as output from nipype). returns numpy ndarray.
    """
    with open(textfile, 'rb') as f:
        reader = csv.reader(f)
        data = [r[0].split() for r in reader]
        str_coords = [d[-3:] for d in data[1:]]
        lcmax_coords = np.array([[int(value) for value in triplet] for triplet in str_coords])
        return lcmax_coords
```

```python
lcmax_coords = get_localmaxima(localmax)
```

Convert voxel coordinates to MNI coordinates

```python
def vox2mnicoords(coords):
    """
    transforms list of voxel idk coordinates into mni.
    """
    # affine matrix for standard space
    mni_affine = datasets.load_mni152_template().affine
    # apply affine to voxel idk to get mni coords
    coords_mni = np.array([apply_affine(mni_affine, coord) for coord in coords])
    return coords_mni
```

```python
lcmax_mni = vox2mnicoords(lcmax_coords)
```

# Simple main effects

For the intercept of the simple main effects (representing the average effects of familiar or unfamiliar faces), we find large portions of the brain showing above threshold clusters of activation (e.g. almost the entire occipital cortex). This was expected since they don't contain any meaningful comparison/contrast with a control condition. Hence, they are not plotted here.

We also have *no* significant clusters for the positive slopes of the familiar and unfamiliar conditions, but only for the negative slopes (reflecting habituation, i.e. a decrease over the course of the experiment). 

An interesting exception is the reversed contrast of the intercept of unfamiliar faces, which should indicate where there is signal-decrease in response to unfamiliar faces. For the reversed intercept of familiar faces, there were no significant clusters.

## unfamiliar - reversed intercept

```python
def plotinrows(niftifile, contourfile=None, colormap=my_blue_r,
               bottomlim=-20, toplim=50,
               nr_rows=2, zmax=None):
    """Volumetric plot with contours (optional) in rows and columns"""

    # initiate subplots
    fig, axes = plt.subplots(nrows=nr_rows)
    rows = {}
    idx = 0
    
    # loop over slices
    for slc in range(bottomlim,toplim-9,25): # slices at beginning of row
        interval = range(slc, slc+21, 5) # slices withing a row
        
        # background
        rows[idx] = plot_anat(display_mode='z', cut_coords=interval, draw_cross=False, axes=axes[idx])
        
        # add contouors
        if contourfile:
            rows[idx].add_contours(contourfile, colors=['red'], linewidths=.7, contours=1, levels=[0])
        
        # stats map
        if not zmax:
            rows[idx].add_overlay(niftifile, cmap=colormap, colorbar=True)
        else:
            rows[idx].add_overlay(niftifile, cmap=colormap, colorbar=True, vmax=zmax)            
        
        idx += 1
    return fig, axs
```

### create spheres and add them to mni mask

```python
def roispheres(center_coords, roi_values, outfile, rad=12):
    """                 
    Make a nifi image in mni standard space containing ROIs
    with center coordinates and fill values of your liking.
    """
    
    # because that's how many voxels there are in MNI152_2mm
    empty = np.zeros([1, 902629])

    # 'create_spheres' doesn't natively give you the option to
    # assign different values to rois. So we have to work around that.
    for center, roi_value in zip(center_coords, roi_values):
        # create an image with a sphere
        sphere_img = fmri_dataset(create_sphere(coordinates=[center], radius=[12]))
        # populate our initial array
        empty[(sphere_img.samples == 1) & (empty == 0)] += roi_value
        # TODO: empty.astype('int')
    
    # pass array to pymvpa object
    sphere_img.samples = empty
    
    # save
    img = map2nifti(sphere_img)
    img.to_filename(outfile)
```

# Get ROI labels from csv file and append by my new ones

```python
def getlabels(oldlabels_csv, addlabels_list=False):
    """
    read csv file containing ROI labels. Return list of tuples with
    roi values and label names. Optionally append existing label list.
    """
    with open(oldlabels_csv, 'r') as f:
        rdr = csv.reader(f)
        content = [row for row in rdr]

    old_labelnames = [row[0] for row in content[1:]]
    old_labels = [(idx, l) for (idx, l) in enumerate(old_labelnames, 1)]
    
    if addlabels_list:
        labels = old_labels + addlabels_list
        return labels, old_labels
    else:
        return old_labels
```

# Scatter plot with ellipsoids representing rois

load statistical map with roi labels

```python
def load_with_labels(nifti_img, mask, labels):
    """
    Load 3D nifti image (e.g. a statistical map) and add feature attributes
    for ROI indices and labels derived from mask image and list of labels.
    """

    # remove redundant time dimension from roi mask
    mean_mask = mean_img(load_img(mask))
    
    # load nifti as dataset with mask for roi values
    dataset_roi = fmri_dataset(nifti_img, add_fa={'roi_value' : mean_mask})
    
    # add label names as feature attribute
    labels_map = dict(labels); labels_map[0.0] = ''
    dataset_roi.fa['roi_label'] = [labels_map[v] for v in dataset_roi.fa.roi_value]
    
    return dataset_roi
```

use this functionality to plot with ellipsoids

```python
def scatter_ellipsoids(nifti1, nifti2, mask, labels, plot_selection=False,
                       pointsize=0.8, outside_roi=False, grid=False,
                       axnames=['image1', 'image2'],
                       thresholds=False, iline=False,
                       colormap='jet'):
    """
    scatter plot two images and add ellipsoids for roi labels.
    provide a mask with roi indices (as integers) and a list of tuples
    mapping the roi indices to labels.
    """
    
    # decide if draw grid
    if grid:
        sns.set_style('whitegrid')
    else:
        sns.set_style('white')
    
    # load both niftis and add roi labels as feature attributes
    ds1 = load_with_labels(nifti1, mask, labels)
    ds2 = load_with_labels(nifti2, mask, labels)
    
    fig = pl.figure(figsize=(14,10))
    ax1 = pl.axes(aspect=1)

    # decide wether to plot all rois or just given selection
    if plot_selection:
        rois = np.array(plot_selection)[:,1]
    else:
        rois = np.unique(ds1.fa.roi_label)
    
    # make color range
    cmap = plt.get_cmap(colormap)
    colors = cmap(np.linspace(0, 1.0, len(rois)))
    
    # assign random colors (rgb) to roi labels
    roicolors = [{'roi' : rl,
                  'rgb' : (random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1))}
                  for rl in rois if rl != '']
    
    # iterate over rois
    for rl, co in zip(rois, colors):
        
        # plot non-roi voxels in grey, if specified
        if rl == '':
            if outside_roi:
                ax1.scatter(x=ds1.samples[0][ds1.fa.roi_label==rl],
                            y=ds2.samples[0][ds2.fa.roi_label==rl],
                            color='grey', label=rl, s=pointsize)
                continue
            else:
                continue

        else:
            #import pdb; pdb.set_trace()
            # find color in defined roi colors
            roicolor = (colordict['rgb'] for colordict in roicolors if colordict['roi'] == rl).next()

            # scatter voxels with certain roi label
            ax1.scatter(x=ds1.samples[0][ds1.fa.roi_label==rl],
                        y=ds2.samples[0][ds2.fa.roi_label==rl],
                        color=co, label=rl, s=pointsize)

            # assemble nd array with z values to get mean and covariance matrix from
            z_array = np.array([ds1.samples[0][ds1.fa.roi_label==rl],
                                ds2.samples[0][ds2.fa.roi_label==rl]])

            # add ellipsoid
            _make_ellipse(z_array.mean(1)[:2],
                              np.cov(z_array)[:2, :2],
                              ax1,
                              color=co)
    # axis labels
    plt.xlabel(axnames[0], fontweight='bold', fontsize=12)
    plt.ylabel(axnames[1], fontweight='bold', fontsize=12)
    
    # threshold lines
    if thresholds:
        plt.axhline(y=thresholds[0], linestyle='--', color='gray')
        plt.axvline(x=thresholds[1], linestyle='--', color='gray')
 
    # identity line
    if iline:
        iline = Line2D(range(-3, 6), range(-3, 6), linestyle='--', color='gray')
        ax1.add_line(iline)

    # plot legend
    ax1.legend(markerscale=10, frameon=True, edgecolor='black', bbox_to_anchor=(1, 1))
    plt.show()

    return fig, ax1
```
