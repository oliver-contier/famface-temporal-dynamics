# Plots for the 2nd level regressor coding a linear decrease of the contrast familiar > unfamiliar over the course of the experiment

```python
# show matplotlib figures in notebook instead of popup window
%matplotlib inline
```

```python
from nilearn.plotting import plot_stat_map
from mvpa2.misc.plot.lightbox import plot_lightbox
import os
from os.path import join
import csv
import matplotlib.pyplot as plt
from urllib import urlretrieve
```

```python

# specify file paths

basedir = '/data/famface/openfmri/oli/results/results_bothregressors/l2ants_fwhm6_hp60_derivs_frac0.1/model001/task001/subjects_all/stats'
contrast = 'contrast__l1-03-l2-02'
localmax = join(basedir, contrast, 'zstat1_reversed_localmax.txt')
filename = join(basedir, contrast, 'zstat1_reversed_threshold.nii.gz')
```

```python

# get coordinates of localmaxima from file

with open(localmax, 'rb') as f:
    reader = csv.reader(f)
    data = [r[0].split() for r in reader]
    str_coords = [tuple(d[-3:]) for d in data[1:]]
    # cast from string to int
    lcmax_coords = [[int(value) for value in triplet] for triplet in str_coords]

    print(lcmax_coords)
```

##  # PyMVPA lightbox plot

```python
# some settings for the plot

# path to mni template
mni_bg = '/usr/share/fsl/data/standard/MNI152_T1_2mm_brain.nii.gz'

# chose slices to plot manually
"""
kmin = 23
kmax = 58
kstep = 3
slc= range(kmin, kmax+1, kstep)
"""

# chose slices that contain local maxima
slc = [i[-1] for i in lcmax_coords]

print(slc)

# color map
# cm = 'Blues'
```

```python

# plot figure

fig = plot_lightbox(background=mni_bg, overlay=filename, slices=slc, ncolumns=4) #cmap_overlay=cm 
```
