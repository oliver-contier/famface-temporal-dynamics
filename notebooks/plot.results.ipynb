```python
# show matplotlib figures in notebook instead of popup window
%matplotlib inline
import numpy as np
from mvpa2.datasets.mri import fmri_dataset
from nilearn.plotting import plot_stat_map, plot_img, plot_prob_atlas
from nilearn.image import load_img, math_img, mean_img, index_img
from mvpa2.misc.plot.lightbox import plot_lightbox
from mvpa2.misc.plot.scatter import plot_scatter
import os
from os.path import join
import csv
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# specify file paths

basedir = '/data/famface/openfmri/oli/results/results_bothregressors/l2ants_fwhm6_hp60_derivs_frac0.1/model001/task001/subjects_all/stats'
contrast = 'contrast__l1-03-l2-02'
localmax = join(basedir, contrast, 'zstat1_reversed_localmax.txt')
filename = join(basedir, contrast, 'zstat1_reversed_threshold.nii.gz')
```

```python
# get coordinates of local maxima from file

def get_localmaxima(textfile):
    with open(textfile, 'rb') as f:
        reader = csv.reader(f)
        data = [r[0].split() for r in reader]
        str_coords = [tuple(d[-3:]) for d in data[1:]]
        # cast from string to int
        lcmax_coords = [[int(value) for value in triplet] for triplet in str_coords]
        return lcmax_coords

lcmax_coords = get_localmaxima(localmax)
print(lcmax_coords)
```

## plot clusters of both linear and average regressors in one volume

red: average effect of familiar > unfamiliar
blue: linear decrease of familiar > unfamiliar

```python
# pseudo-binarize linear effect
linear_img = load_img(filename)
linear_img_ms = math_img('img != 0', img=linear_img)

# pseudo-binarize average effect
average_img = load_img(intc_filename)
average_img_ms = math_img('img != 0', img=average_img)

# plot both overlays in nilearn
plot = plot_stat_map(average_img_ms, display_mode='z', cut_coords=10, draw_cross=False,
                     cmap='coolwarm', colorbar=False)
plot.add_overlay(linear_img_ms, colorbar=False)

#cut_coords=(0,-28,32)
```
