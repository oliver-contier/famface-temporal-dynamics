```python
# get coordinates of local maxima from file
def get_localmaxima(textfile):
    """
    Reads csv-file with local maxima (as output from nipype). returns numpy ndarray.
    """
    with open(textfile, 'rb') as f:
        reader = csv.reader(f)
        data = [r[0].split() for r in reader]
        str_coords = [d[-3:] for d in data[1:]]
        lcmax_coords = np.array([[int(value) for value in triplet] for triplet in str_coords])
        return lcmax_coords
```

```python
lcmax_coords = get_localmaxima(localmax)
```

```python
def vox2mnicoords(coords):
    """
    transforms list of voxel idk coordinates into mni.
    """
    # affine matrix for standard space
    mni_affine = datasets.load_mni152_template().affine
    # apply affine to voxel idk to get mni coords
    coords_mni = np.array([apply_affine(mni_affine, coord) for coord in coords])
    return coords_mni
```

```python
lcmax_mni = vox2mnicoords(lcmax_coords)
```

# Create spherical ROIs around local maxima

```python
# my selection of coordinates for roi centers
lcmax_select_idk = [(45,42,49), (45,50,52), (47,59,52), (32,91,48), (21,70,47), (27,70,42), (25,68,34), (20, 38, 26)]
                                                           
# converted to mni coords
lcmax_select_array = vox2mnicoords(lcmax_select_idk)
    
# but stupid 'create_sphere' function wants them as list of tuples, so... 
lcmax_select = [tuple(i) for i in lcmax_select_array]
```

```python
def roispheres(center_coords, roi_values, outfile, rad=12):
    """                 
    Make a nifi image in mni standard space containing ROIs
    with center coordinates and fill values of your liking.
    """
    
    # because that's how many voxels there are in MNI152_2mm
    empty = np.zeros([1, 902629])

    # to frown upon: 'create_spheres' doesn't natively give you the option
    # to assign different values to rois. So we have to work around that.
    for center, roi_value in zip(center_coords, roi_values):
        # create an image with a sphere
        sphere_img = fmri_dataset(create_sphere(coordinates=[center], radius=[12]))
        # populate our initial array
        empty[(sphere_img.samples == 1) & (empty == 0)] += roi_value
    
    # pass array to pymvpa object
    sphere_img.samples = empty
    
    # save
    nib = map2nifti(sphere_img)
    nib.to_filename(outfile)
```

```python
# apply function to create spheres
outfilename = '/data/famface/openfmri/oli/results/extract_betas/newrois_overlap.nii.gz'
roivalues = range(31,39)
roispheres(lcmax_select, roivalues, outfilename)
```

- the resulting image contains spherical ROIs based on the local maxima from our contrast of interest
- although they do not overlap, the ROI values were assigned based on "first come first serve"
- I manually assigned these ROIs new borders to equalize size and in order for them not to overlap with Matteo's ROIs
