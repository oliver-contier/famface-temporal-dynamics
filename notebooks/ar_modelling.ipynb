```python
%matplotlib inline
from statsmodels.tsa.ar_model import AR
import pandas as pd
import os
from mvpa2.suite import *
import matplotlib.pyplot as plt
```

# Plot time series and residuals from AR

```python
# read csv into data frame
run001csv = '/data/famface/openfmri/oli/results/extract_meants/csv/run001/sub001_run001.csv'
df_run001 = pd.read_csv(run001csv)
```

```python
# add columns for time in seconds, and create a time stamp based on that (because needed by statsmodels).
df_run001['time'] = range(2,154*2+1,2)
df_run001.index = pd.to_datetime(pd.to_timedelta(df_run001['time'],unit='s'), dayfirst=True, yearfirst=True)
```

perform auto regressive modelling for each of those rois and plot residuals vs actual data

```python
def plot_datavsres(df, cond):
    """
    data frame must have index in datetime format and a variable 'time'.
    """
    plt.figure()
    # set up model for one roi
    ar_mod = AR(df[cond].astype(float), dates=df.index)
    # plot data
    plt.plot(df['time'], df[cond])
    # for different lags
    for lag in range(1,5):
        # fit ar model
        ar_res = ar_mod.fit(1)
        res = ar_res.predict()
        # plot residuals
        plt.plot(df['time'][1:], res)
    plt.title(cond)
    plt.legend(['data', '1','2','3','4'])
    plt.show()
```

```python
# for one ROI
plot_datavsres(df_run001, 'EV1-L')
```

so indeed, the different lags don't seem to make much of a difference.

## make use of Yarik's function for plotting i.i.d of our data

```python
def plot_corrs(data, offsets = list(range(20))):
    # should be 1 for the identical signal (offset = 0)
    dcorrs = [dcorcoef(data[o:].T, data[:-o or len(data)].T) for o in offsets]
    pl.plot(offsets, dcorrs, label='dcorcoef' + (str(data.shape[-1] if data.ndim > 1 else ''))); pl.xlabel("offset")
    corrs = None
    if data.ndim == 1:
        # and what about auto-correlation
        corrs = [np.corrcoef(data[o:], data[:-o or len(data)])[0, 1] for o in offsets]
        pl.plot(offsets, corrs, label='corrcoef');
    pl.hlines(0, 0, max(offsets), color='g', linestyles='--')
    pl.legend()
    return dcorrs, corrs
```

```python
# plot for each roi
for roi in df_run001.columns:
    if roi != 'time':
        pl.figure()
        ddc, dcc = plot_corrs(df_run001[roi])
        pl.title(roi)
```
