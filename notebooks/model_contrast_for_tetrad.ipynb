This notebook is for visualization of how the familiarity contrast is modelled as the difference between the two HRF models for familiar and unfamiliar faces.

```python
from os import chdir, getcwd
import os.path
from os.path import join as pjoin
from matplotlib import pyplot as plt
import numpy as np
from mvpa2.misc.data_generators import simple_hrf_dataset
```

```python
%matplotlib inline
```

```python
# import function for getting simulus onsets as used during simulation
chdir('/data/famface/openfmri/oli/osf_prereg_code/simulation')
from famface_simulation_functions import get_onsets_famface
```

```python
# path to onset files for sub001, first run
basedir = '/data/famface/openfmri/oli/simulation/data_oli'
sub = 'sub001'
run = 'task001_run001'
inpath = pjoin(basedir, sub, 'model/model001/onsets', run)
```

```python
amplitudes = [1,1]
spec = get_onsets_famface(inpath, amplitudes)
```

```python
print(spec[0]['targets'])
print(spec[0]['onset'])
```

```python
hrf_models = []

for condition in spec:
    hrf_model = simple_hrf_dataset(events=condition['onset'],
                          nsamples=154,
                          tr=2,
                          tres=2,
                          baseline=0,
                          signal_level=amplitudes[0],
                          noise_level=0)
    hrf_models.append(hrf_model)
```

```python
for hrf_model in hrf_models:
    plt.plot(hrf_model.samples[:,0])
    
plt.legend(['familiar', 'unfamiliar'], loc='upper left')
plt.xlabel('TR')
plt.title('HRF models per condition', fontweight='bold')
```

```python
fam_vs_unfam = hrf_models[0].samples[:,0] - hrf_models[1].samples[:,0]
np.shape(fam_vs_unfam)
```

```python
plt.plot(fam_vs_unfam, 'red')
plt.xlabel('TR')
plt.title('Differential Regressor: FAM - UNFAM', fontweight='bold')
```

So in conclusion, this is how we construct a variable that reflects the difference between the HRF models for the presented familiar and unfamiliar faces respectively. This variable will be used to model the experimental manipulation in our py-causal analysis.
