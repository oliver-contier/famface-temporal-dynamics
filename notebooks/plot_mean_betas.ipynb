```python
%matplotlib inline
from mvpa2.datasets.mri import fmri_dataset
from mvpa2.misc.plot.scatter import plot_scatter
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
import pandas as pd
import numpy as np
from os.path import join
import os
import csv
import sys
```

```python
def append_sub(csvfile, sub, ptype):
    """
    append data from one subject to a data frame
    """
    
    df = pd.DataFrame()
    
    # read from csv
    csv_df = pd.read_csv(csvfile, header=None)

    # iterate over the 3 clusters
    for c in list(csv_df.columns.values):

        data = {'parameter_estimate': csv_df[c],
                'roi': (list(csv_df.columns.values).index(c)+1),
                'run': range(1, len(csv_df)+1),
                'subject': sub,
                'stats': ptype}

        # append to resulting data frame
        temp_df = pd.DataFrame(data)
        df = df.append(temp_df, ignore_index=True)
    
    result_df = df
    return result_df
```

```python
def extract_all(parameters, parameter_types, outdir):
    """
    extract all data from csv files into pandas, given parameters,
    parameter types, and output directory of the csv files.
    """

    final_df = pd.DataFrame()
    outdir = join('/data/famface/openfmri/oli/results/extract_betas/', outdir)

    # append for each condition and each subject
    for ptype in parameter_types:
        for parameter in parameters:
            for sub in range(1, 34):
                # call function defined above
                csvfile = join(outdir, 'sub%03d' % sub, '%s_mean_%s_sub%03d.csv' % (parameter, ptype, sub))
                df_with_sub = append_sub(csvfile, sub, ptype)
                df_with_sub['parameter'] = parameter
                final_df = final_df.append(df_with_sub, ignore_index=True)

    return final_df
```

```python
def doubleplot(df, roilabel=None, stat='cope'):
    """
    Plot the progression of parameter estimates over runs in two columns:
    One for the simple regressors and one for the contrast.
    """

    # initiate subplots
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False, figsize=(15,6))
    
    # if no roilabel given, plot mean over all data
    if roilabel==None:
        # plot two columns (one for both regressors and one for contrast)
        c1 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax1, palette='colorblind',
                            data=df[df.stats==stat][(df.parameter == 'familiar') | (df.parameter=='unfamiliar')],
                            kind='point',
                            ci=95)
        c2 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax2, palette=ocred,
                            data=df[df.stats==stat][df.parameter == 'famvsunfam'],
                            kind='point',
                            ci=95)

    else:
        c1 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax1, palette='colorblind',
                                data=df[(df.stats==stat) & (df.label==roilabel)][(df.parameter == 'familiar') | (df.parameter=='unfamiliar')],
                                kind='point',
                                ci=95)
        c2 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax2, palette=ocred,
                                data=df[(df.stats==stat) & (df.label==roilabel)][df.parameter == 'famvsunfam'],
                                kind='point',
                                ci=95)
        
        # make roi titles
        ax1.title.set_text(roilabel)
        ax1.title.set_position(xy=[0, 1])
        ax1.title.set_fontproperties(title_font)
        
    # y axis label
    for ax in [ax1, ax2]:
        ax.set_ylabel('ROI mean %s' % stat)

    # close redundent subplots, that are produced for unknown reasons...
    plt.close(c1.fig)
    plt.close(c2.fig)
```

create my own color palette for convenience

```python
ocred=sns.hls_palette(8, l=.5, s=0.6)
```

create custom font for roi titles

```python
default_font = FontProperties()
title_font = default_font.copy()
title_font.set_weight('bold')
title_font.set_size(14)
```

## get roi labels from Matteos mask and extend by mine

```python
def getlabels(csvpath):
    with open(csvpath, 'r') as f:
        rdr = csv.reader(f)
        content = [row for row in rdr]

    labelnames = [row[0] for row in content[1:]]
    labels = [(idx, l) for (idx, l) in enumerate(labelnames, 1)]
    return labels
```

# Seperately for the 3 Clusters we found

```python
for roi in sorted(set(df_cl.roi.values)):

    # initiating subplots
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False, figsize=(15,6))

    # plot two columns
    c1 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax1, palette='colorblind',
                        data=df_cl[(df_cl.stats=='cope') & (df_cl.roi==roi)][(df_cl.parameter == 'familiar') | (df_cl.parameter=='unfamiliar')],
                        kind='point',
                        ci=95)

    c2 = sns.factorplot('run', hue='parameter', y='parameter_estimate', ax=ax2, palette=ocred,
                        data=df_cl[(df_cl.stats=='cope') & (df_cl.roi==roi)][df_cl.parameter == 'famvsunfam'],
                        kind='point',
                        ci=95)
    
    # add roi number/name in title
    ax1.title.set_text('cluster %02d ' % roi)
    ax1.title.set_position(xy=[0, 1])
    ax1.title.set_fontproperties(title_font)
    
    # close redundent subplots, that are produced for unknown reasons...
    plt.close(c1.fig)
    plt.close(c2.fig)
```

- familiar > unfamiliar seems to be above 0 only in the first 3 runs.
- in the last 3 runs, it even seems to be negative.
- (the variance of the contrast increases over runs, which is expected since it relies on the variance of the regressors)
