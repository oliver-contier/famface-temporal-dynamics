```python
def append_sub(csvfile, sub, ptype):
    """
    Read csv with mean parameter estimates for one subject
    and append data to data frame.
    """
    
    df = pd.DataFrame()
    
    # read from csv
    csv_df = pd.read_csv(csvfile, header=None)

    # iterate over the 3 clusters
    for c in list(csv_df.columns.values):

        data = {'parameter_estimate': csv_df[c],
                'roi': (list(csv_df.columns.values).index(c)+1),
                'run': range(1, len(csv_df)+1),
                'subject': sub,
                'stats': ptype}

        # append to resulting data frame
        temp_df = pd.DataFrame(data)
        df = df.append(temp_df, ignore_index=True)
    
    result_df = df
    return result_df
```

```python
def extract_all(parameters, parameter_types, outdir):
    """
    extract all data from csv files into pandas, given parameters,
    parameter types, and output directory of the csv files.
    """

    final_df = pd.DataFrame()
    outdir = join('/data/famface/openfmri/oli/results/extract_betas/', outdir)

    # append for each condition and each subject
    for ptype in parameter_types:
        for parameter in parameters:
            for sub in range(1, 34):
                # call function defined above
                csvfile = join(outdir, 'sub%03d' % sub, '%s_mean_%s_sub%03d.csv' % (parameter, ptype, sub))
                df_with_sub = append_sub(csvfile, sub, ptype)
                df_with_sub['parameter'] = parameter
                final_df = final_df.append(df_with_sub, ignore_index=True)

    return final_df
```

```python
def doubleplot(df, roilabel=False, stat='cope', plottype='point', hue='parameter',
               fitlines=False, subcolors=False, zoomout=False):
    """
    Plot the progression of parameter estimates over runs in two columns:
    One for the simple regressors and one for the contrast.
    """
    nr_runs=11
    sns.set_style("whitegrid")

    # initiate subplots
    fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False, figsize=(15,6))
    
    # if no roilabel given, plot mean over all data
    if not roilabel:
        # plot two columns (one for both regressors and one for contrast)
        c1 = sns.factorplot('run', hue=hue, y='parameter_estimate', ax=ax1, palette='colorblind',
                            data=df[df.stats==stat][(df.parameter == 'familiar') | (df.parameter=='unfamiliar')],
                            kind=plottype,
                            ci=95)
        c2 = sns.factorplot('run', hue=hue, y='parameter_estimate', ax=ax2, palette=ocred,
                            data=df[df.stats==stat][df.parameter == 'famvsunfam'],
                            kind=plottype,
                            ci=95)

    else:
        c1 = sns.factorplot('run', hue=hue, y='parameter_estimate', ax=ax1, palette='colorblind',
                                data=df[(df.stats==stat) & (df.label==roilabel)][(df.parameter == 'familiar') | (df.parameter=='unfamiliar')],
                                kind=plottype,
                                ci=95)
        c2 = sns.factorplot('run', hue=hue, y='parameter_estimate', ax=ax2, palette=ocred,
                                data=df[(df.stats==stat) & (df.label==roilabel)][df.parameter == 'famvsunfam'],
                                kind=plottype, zorder=1,
                                ci=95)
        
        if fitlines:
            # look for matching fit lines, add to plot, and store in list (for mean fit line)
            matched_lines = []
            for linedict in fitlines:
                if linedict['roilabel'] == roilabel and linedict['stat'] == stat and linedict['cope'] == 'famvsunfam':
                    
                    #if each subject should have its own color
                    if subcolors:
                        for subcolor in subcolors:
                            if subcolor['sub'] == linedict['subject']:
                                line_obj = Line2D(range(nr_runs), linedict['line'],
                                                  c=subcolor['rgb'], linestyle='-', zorder=0)
                    else:
                        line_obj = Line2D(range(nr_runs), linedict['line'],
                                                  c='grey', linestyle='-', zorder=0, alpha=0.6)
                    ax2.add_line(line_obj)
                    matched_lines.append(np.array(linedict['line']))


            # add a mean fit line
            meanline = np.mean(matched_lines, 0)
            meanline_obj = Line2D(range(nr_runs), meanline,
                                  c='black', linestyle='--', zorder=3)
            ax2.add_line(meanline_obj)
        
        # make roi titles
        ax1.title.set_text(roilabel)
        ax1.title.set_position(xy=[0, 1])
        ax1.title.set_fontproperties(title_font)
    
    # y axis label
    for ax in [ax1, ax2]:
        ax.set_ylabel('ROI mean %s' % stat)

    # zoom out if wanted
    if zoomout:
        ax2.set_ylim(zoomout)
    
    # close redundent subplots, that are produced for unknown reasons...
    plt.close(c1.fig)
    plt.close(c2.fig)
```

```python
def makesubcolors(N):
    """
    assign each of N subjects a random RGB value.
    """

    subcolors = [{'sub' : sub,
                  'rgb' : (random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1))}
                 for sub in range(1,N+1)]

    return subcolors
```

```python
def getlabels(oldlabels_csv, addlabels_list=False):
    """
    read csv file containing ROI labels. Return list of tuples with
    roi values and label names. Optionally append existing label list.
    """
    with open(oldlabels_csv, 'r') as f:
        rdr = csv.reader(f)
        content = [row for row in rdr]

    old_labelnames = [row[0] for row in content[1:]]
    old_labels = [(idx, l) for (idx, l) in enumerate(old_labelnames, 1)]
    
    if addlabels_list:
        labels = old_labels + addlabels_list
        return labels, old_labels
    else:
        return old_labels
```

## Fitline from 2nd lvl results

Calculate a fitline for each subject based on the intercept and slope images of our second level analysis.

```python
# import function we defined for extracting parameter estimates on the first level
sys.path.insert(0, '/data/famface/openfmri/oli/osf_prereg_code/analysis')
from extract_pe import extract_mean_3d
```

First we define a function for extracting the mean ROI values for one subject. Then another one to iterate over all kinds of subjects, statistics (z-stats, copes, varcopes) and first level contrasts (familiar, unfamiliar, fam>unfam, unfam>fam).

```python
def getfit(intc, slp, mask, sub, stat, cope, nr_runs, labels):
    """
    create fitlines for all rois in a mask given
    intercept and slope 2nd lvl statistical map
    (i.e. cope images) for one subject.
    
    returns list of dicts with line objects and information
    about subject id, stats type, and roi value.
    """
    
    # load images in pymvpa
    intc_img = fmri_dataset(intc)
    slp_img = fmri_dataset(slp)
    ms = fmri_dataset(mask)
    
    # find out number of rois in mask image
    nr_rois = int(max(ms.samples[0]))
    
    # extract mean parameter estimates for each roi
    intercepts = extract_mean_3d(intc_img, ms)
    slopes = extract_mean_3d(slp_img, ms)
    
    # create fitlines for each roi, returned as list of dicts
    fitlines = []
    for roi in range(1, nr_rois+1):
        fitline = [intercepts[roi-1] + (slopes[roi-1] * run) for run in range(nr_runs)]
        fitlines.append(
            {'subject' : sub,
             'roi_value' : roi,
             'line' : fitline,
             'stat' : stat,
             'cope' : cope, 
             'roilabel' : np.array(labels)[:,1][roi-1]})
    
    return fitlines
```

```python
def getallfits(basedir, mask, labels,
               stats=['zstat', 'cope', 'varcope'],
               copes=['familiar', 'unfamiliar', 'famvsunfam', 'unfamvsfam'],
               nruns=11):
    """
    execute getfit for all subjects, copes, and stats.
    """
    
    fitlines = []
    for sub in range(1,34):
        for cope in copes:
            copenr = copes.index(cope)+1
            for stat in stats:
                intc = join(basedir, 'sub%03d'%sub, stat+'s', 'mni', '%s_l1-%02d-l2-01.nii.gz' % (stat, copenr))
                slp = join(basedir, 'sub%03d'%sub, stat+'s', 'mni', '%s_l1-%02d-l2-02.nii.gz' % (stat, copenr))
                temp_fitlines = getfit(intc, slp, mask, sub, stat, cope, nr_runs, labels)
                fitlines.append(temp_fitlines)
    
    # flatten the list of lists
    fitlines = [item for sublist in fitlines for item in sublist]
    
    return fitlines
```

This step executes the above definitions and takes a while to run (which is why we save the result as a pickle file below, so we don't have to run again that often).

## Varcope

FSL's "varcope" output images reflect the within-run variance of the parameter estimates. They are used for calculation of T-Scores ( T = cope / sqrt[varcope] )

Here we plot the average varcope over all 3 clusters from our final analysis.

```python
doubleplot(df_cl, stat='varcope', plottype='point')
```

We see an increase of the variance in parameter estimates with run number.

```python
doubleplot(df_cl, stat='varcope', plottype='box')
```

Looking at boxplots instead of just the mean and CIs, it appears that this increase is more driven by outlier subjects and a widening of the distributions than actual increases of the means.

** Conclusions **

- familiar > unfamiliar seems to be above 0 only in the first 3 runs.
- in the last 3 runs, it even seems to be negative.
- (the variance of the contrast increases over runs, which is expected since it relies on the variance of the regressors)
