```python
%matplotlib inline
from mvpa2.datasets.mri import fmri_dataset
from mvpa2.misc.plot.scatter import plot_scatter
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from os.path import join
import os
```

```python
# append data from one subject to a data frame
def append_sub(csvfile, sub, ptype):
    
    df = pd.DataFrame()
    
    # read from csv
    csv_df = pd.read_csv(csvfile, names=['cluster1', 'cluster2', 'cluster3'])

    # iterate over the 3 clusters
    for c in list(csv_df.columns.values):

        data = {'cluster_mean': csv_df[c],
                'cluster': (list(csv_df.columns.values).index(c)+1),
                'run': range(1, len(csv_df)+1),
                'subject': sub,
                'stats': ptype}

        # append to resulting data frame
        temp_df = pd.DataFrame(data)
        df = df.append(temp_df, ignore_index=True)
    
    result_df = df
    return result_df
```

```python
def extract_all(parameters, parameter_types):

    # initialize data frame
    final_df = pd.DataFrame()

    # directory with csv files
    outdir = '/data/famface/openfmri/oli/results/extract_betas/outdir'

    # append for each condition and each subject
    for ptype in parameter_types:
        for parameter in parameters:
            for sub in range(1, 34):
                csvfile = join(outdir, 'sub%03d' % sub, '%s_mean_%s_sub%03d.csv' % (parameter, ptype, sub))
                df_with_sub = append_sub(csvfile, sub, ptype)
                df_with_sub['parameter'] = parameter
                final_df = final_df.append(df_with_sub, ignore_index=True)

    return final_df
```

```python
df_all = extract_all(
    ['familiar', 'unfamiliar', 'famvsunfam', 'unfamvsfam'],
    ['tstat', 'zstat', 'cope', 'varcope']
)
```

# Plot each type of statistics seperately (pe, zstat, cope, varcope) for all conditions, averaged over clusters

To get a rough idea if the conditions / parameters differ in their magnitude.

```python
# average over all clusters

for stat in ['tstat', 'zstat', 'cope', 'varcope']:
    ax_av = sns.factorplot('run', hue='parameter', y='cluster_mean',
                        data=df_all[(df_all.stats==stat)][(df_all.parameter == 'familiar') | (df_all.parameter=='unfamiliar')],
                        kind='point', size=7)
```

```python
# clusters in columns

for stat in ['tstat', 'zstat', 'cope', 'varcope']:
    ax_cl = sns.factorplot('run', hue='parameter', y='cluster_mean',
                        data=df_all[(df_all.stats==stat)][(df_all.parameter == 'familiar') | (df_all.parameter=='unfamiliar')],
                        kind='point', size=7, col='cluster')
```
